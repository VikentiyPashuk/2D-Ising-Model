```python
import numpy as np
import matplotlib.pyplot as plt

# Read data from the text files
spins_initial_matrix = np.loadtxt('spins_initial_matrix.txt')
spins_final_matrix = np.loadtxt('spins_final_matrix.txt')

# Create figure and axis objects
fig, axes = plt.subplots(1, 2, figsize=(10, 8))

# Plot spins_matrix_up
axes[0].imshow(spins_initial_matrix, cmap='gray', origin='lower', extent=[0, 10, 0, 10])
axes[0].set_title('a) Initial Spins at Low Temperature', fontweight='bold', fontsize=16)
axes[0].set_xticks(np.arange(0, 11, 1))
axes[0].set_yticks(np.arange(0, 11, 1))
axes[0].grid(which='both', color='black', linestyle='-', linewidth=1)

# Plot spins_matrix_down
axes[1].imshow(spins_final_matrix, cmap='gray', origin='lower', extent=[0, 10, 0, 10])
axes[1].set_title('b) Final Spins at High Temperature', fontweight='bold', fontsize=16)
axes[1].set_xticks(np.arange(0, 11, 1))
axes[1].set_yticks(np.arange(0, 11, 1))
axes[1].grid(which='both', color='black', linestyle='-', linewidth=1)

# Replace positive and negative values with colored boxes
for ax, spins_matrix in zip(axes, [spins_initial_matrix, spins_final_matrix]):
    for i in range(spins_matrix.shape[0]):
        for j in range(spins_matrix.shape[1]):
            if spins_matrix[i, j] == 1:
                ax.add_patch(plt.Rectangle((j, i), 1, 1, color='yellow'))
            elif spins_matrix[i, j] == -1:
                ax.add_patch(plt.Rectangle((j, i), 1, 1, color='blue'))

# Add text box introducing the legend to the right of the most right plot
legend_text = 'Legend:\nUp Spin - Yellow Square \nDown Spin - Blue Square'
fig.text(1.025, 0.5, legend_text, fontsize=12, va='center', color='black', bbox=dict(facecolor='white', edgecolor='black'))

# Adjust layout and save the plot
plt.tight_layout()
plt.savefig('spins_grid.pdf', dpi=300, bbox_inches='tight')  # Save the plot as a PDF file
plt.show()
```
